##We can see that most empty values are associated with goalkeeping statistics as number of goalkeepers are less compared to outfield players

```python
#descriptive statistics of dataset
data.describe()
```

```python
data['Pos'].value_counts().plot(kind='bar')
plt.title('Top 10 Player Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

##Conclusion: The data is skewed toward defenders and midfielders, which reflects real-world squad building where these positions offer greater tactical flexibility and squad depth.

```python
sns.histplot(data['Age'].dropna(),bins=20,color='teal')
plt.title("Age Distribution of Players")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

##Conclusion: The majority of players are aged between 20 and 28, aligning with the physical and technical peak years in professional football. This suggests the dataset predominantly includes players in their prime.

```python
data['Comp'].value_counts().head(6).plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3"))
plt.title("Top 6 Competitions Represented")
plt.ylabel("")
plt.show()
```

##Conclusion: Serie A, Premier League, and La Liga dominate the player representation, highlighting their competitiveness, media reach, and influence in European football.

```python
data['Squad'].value_counts().head(10).plot(kind='barh', color='lightgreen')
plt.title("Top 10 Squads by Player Count")
plt.xlabel("Number of Players")
plt.ylabel("Squad")
plt.gca().invert_yaxis()
plt.show()
```

##Conclusion: Clubs with larger representations often compete in multiple tournaments and maintain deeper squads to manage intense schedules or rotate between youth and senior players.

```python
top_scorers = data.sort_values('Gls',ascending=False).head(10)
plt.bar(x=top_scorers['Player'],height=top_scorers['Gls'],color='magenta')
plt.title("Top 10 Goal Scorers")
plt.ylabel("Goals")
plt.xticks(rotation=45)
plt.show()
```

##Conclusion:This identifies elite finishers in the dataset, with players like Mbappé leading by a margin, indicating high shot accuracy, positioning, and consistency in goal scoring.

```python
sns.boxplot(x='Pos',y='Gls',data=data)
plt.title("Goals by Position")
plt.xlabel("Position")
plt.ylabel("Goals")
plt.xticks(rotation=45)
plt.show()
```

##Conclusion: The plot confirms expected roles: forwards score significantly more, while defenders contribute fewer goals, supporting tactical role assignments in formations.

```python
top_assisters=data.sort_values('Ast',ascending=False).head(10)
plt.bar(x=top_assisters['Player'], height=top_assisters['Ast'],color='teal')
plt.title("Top 10 Assist Providers")
plt.ylabel("Assists")
plt.xticks(rotation=45)
plt.show()
```

##Conclusion: Playmakers and wide players dominate this list, reinforcing their roles as creative engines for their teams and their efficiency in setting up scoring opportunities.

```python
sns.scatterplot(x='Att',y='Succ',data=data)
plt.title("Dribbles Attempted vs Successful Dribbles")
plt.xlabel("Dribbles Attempted")
plt.ylabel("Successful Dribbles")
plt.show()
```

##Conclusion: There's a strong positive relationship between attempted and successful dribbles, showing that high-volume dribblers typically have both confidence and skill execution.

```python
sns.scatterplot(x='PrgP',y='Ast',data=data)
plt.title("Progressive Passes vs Assists")
plt.xlabel("Progressive Passes")
plt.ylabel("Assists")
plt.show()
```

##Conclusion: Players who make more progressive passes also tend to register more assists, suggesting that verticality in passing is key to playmaking effectiveness.

```python
data['Cards'] = data['CrdY']+data['CrdR']
sns.histplot(data['Cards'],bins=10,color='darkred')
plt.title("Distribution of Total Cards")
plt.xlabel("Cards")
plt.ylabel("Players")
plt.show()
```

##Conclusion: Most players maintain a disciplined game, receiving fewer than 5 cards in a season. A few outliers may indicate aggressive or high-risk playing styles.

```python
data['GA'] = data['Gls'] + data['Ast']
top_ga = data.sort_values('GA', ascending=False).head(10)
plt.bar(x=top_ga['Player'], height=top_ga['GA'], color='darkorange')
plt.title("Top 10 Players by Goal Contributions (Goals + Assists)")
plt.ylabel("Goal Contributions")
plt.xticks(rotation=45)
plt.show()
```

##Conclusion: This metric balances both goals and assists, offering a comprehensive view of attacking influence. Players on this list are central to their teams' scoring output.

```python
gk_df = data[data['Pos'].str.contains('GK', na=False)]
sns.scatterplot(x='Save%', y='Saves', data=gk_df)
plt.title("Saves vs Save Percentage (Goalkeepers)")
plt.xlabel("Save %")
plt.ylabel("Saves")
plt.show()
```

##Conclusion: Goalkeepers facing more shots don’t always have higher save percentages. The plot reveals efficiency under pressure and helps identify shot-stopping specialists.

```python
sns.boxplot(x='Pos', y='Int', data=data)
plt.title("Interceptions by Position")
plt.xlabel("Position")
plt.ylabel("Interceptions")
plt.xticks(rotation=45)
plt.show()
```

##Conclusion: Defensive midfielders and center-backs typically top the chart, reflecting their role in breaking opposition play and regaining possession in defensive zones.

```python
sns.histplot(data['SCA'].dropna(), bins=20, color='blueviolet')
plt.title("Shot Creating Actions (SCA) Distribution")
plt.xlabel("SCA")
plt.ylabel("Players")
plt.show()
```

##Conclusion: Only a subset of players regularly contribute to pre-shot activities. These players often act as the creative link in offensive transitions.

```python
data['GA90'] = data['GA'] / data['90s']
sns.histplot(data['GA90'].dropna(), bins=30, color='mediumturquoise')
plt.title("Goal Contributions per 90 Minutes")
plt.xlabel("GA per 90")
plt.ylabel("Players")
plt.show()
```

##Conclusion: This allows fair comparisons between starters and substitutes. A high GA/90 indicates that even players with fewer minutes can be highly impactful.

```python
sns.violinplot(x='Pos', y='Cmp%', data=data)
plt.title("Pass Accuracy by Position")
plt.xlabel("Position")
plt.ylabel("Pass Completion %")
plt.xticks(rotation=45)
plt.show()
```

##Conclusion: Midfielders show tighter distribution around high accuracy, as they're heavily involved in ball retention. Variance in forwards and defenders reflects riskier pass attempts.

```python
sns.scatterplot(x='Touches', y='KP', data=data)
plt.title("Touches vs Key Passes")
plt.xlabel("Touches")
plt.ylabel("Key Passes")
plt.show()
```

##Conclusion: Players with more touches often make more key passes, especially midfielders who operate as creative hubs. This also points to tempo control responsibilities.

```python
data['xA'] = pd.to_numeric(data['xA'], errors='coerce')
data['xA/90'] = data['xA'] / data['90s']
filtered_df = data[data['90s'] >= 5]
top_xa90 = filtered_df.sort_values('xA/90', ascending=False).head(10)
top_xa90.plot.bar(x='Player', y='xA/90', color='mediumseagreen', legend=False)
plt.title("Top 10 Players by Expected Assists per 90 Minutes (xA/90)")
plt.ylabel("xA per 90")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Conclusion: These players consistently generate high-quality chances regardless of minutes played. It highlights underlying creative talent that raw assist counts may overlook.

```python
data['Overperformance'] = data['Gls'] - data['xG']
top_over = data.sort_values('Overperformance', ascending=False).head(10)
top_over.plot(x='Player', y='Overperformance', kind='barh', color='limegreen')
plt.title('Top 10 Players Who Outperform Their xG')
plt.xlabel('Goals - xG')
plt.gca().invert_yaxis()
plt.show()
```

##Conclusion: This plot highlights players who consistently outperform their expected goals (xG). In other words, it reveals which players are exceptional finishers — those who score more goals than expected based on the quality of their chances.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='1/3', y='Ast', data=data, hue='Pos', palette='Set2')
plt.title("Passes into Final Third vs Assists")
plt.xlabel("Passes into Final Third")
plt.ylabel("Assists")
plt.legend(title="Position", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
